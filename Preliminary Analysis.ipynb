# Preliminary Analysis

### 2021 Statistics Correlation w/ 2022 Yahoo Fantasy Points

```python
import pandas as pd
df2021_2022 = pd.read_csv("2021-2022 Correlation.csv", sep = ",")

# Eliminating String Features to be able to run Correlation Analysis
corr_df2021_2022 = df2021_2022.drop(['Name', 'ID'], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting Correlation HeatMap
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(corr_df2021_2022.corr()[['2022 Yahoo Fantasy Pts']].sort_values(by='2022 Yahoo Fantasy Pts', ascending=False), vmin=-1, vmax=1, 
                      annot=True, cmap='BrBG')
heatmap.set_title('2021 Statistics Correlating with 2022 Yahoo Fantasy Pts', fontdict={'fontsize':18}, pad=16)
```

#### 2021 Statistics vs 2022 Fantasy Points Scatter Plots
(Checking for Non-linear Patterns)

##### 2021 Yahoo Stats Scatter Plots

```python
from pandas.plotting import scatter_matrix

# Basic Yahoo Stats
yahoo21 = ['2022 Yahoo Fantasy Pts', '2021 Yahoo Fantasy Pts', '2021 IP', '2021 Hits', '2021 Strikeouts', '2021 Walks', '2021 Earned Runs', '2021 Wins', '2021 Losses']
scatter_matrix(corr_df2021_2022[yahoo21])
```

##### 2021 Baseball Savant Statcast Scatter Plots

```python
# Statcast Stats
statcast21 = ['2022 Yahoo Fantasy Pts', '2021 xba', '2021 xslg', '2021 woba', '2021 xwoba', '2021 xobp', '2021 xiso', '2021 exit_velocity_avg', '2021 launch_angle_avg', '2021 sweet_spot_percent', '2021 barrel_batted_rate', '2021 hard_hit_percent', '2021 avg_best_speed', '2021 avg_hyper_speed', '2021 whiff_percent', '2021 swing_percent']
scatter_matrix(corr_df2021_2022[statcast21])
```
